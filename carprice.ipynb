```python
train_df=pd.concat([df,newdf],axis=1)
train_df
```

```python
#dropping columns
train_df.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price','Brand'],axis=1,inplace=True)
train_df
```

```python
train_df['Mileage']=train_df['Mileage'].str.replace('km/kg','')
train_df['Mileage']=train_df['Mileage'].str.replace('kmpl','')
```

```python
#removing str values
train_df['Engine']=train_df['Engine'].str.replace('CC','')
train_df['Power']=train_df['Power'].str.replace('bhp','')
train_df
```

```python
train_df['Engine'].unique()
```

```python
train_df['Mileage'].unique()
```

```python
train_df['Power'].unique()
```

```python
#changin null values to zero
train_df['Engine']=train_df['Engine'].str.replace('null','0')
train_df['Mileage']=train_df['Mileage'].str.replace('null','0')
train_df['Power']=train_df['Power'].str.replace('null','0')
train_df
```

```python
train_df.dtypes
```

```python
#changing data types of columns
train_df['Mileage']=train_df['Mileage'].astype(float)
train_df['Engine']=train_df['Engine'].astype(float)
train_df['Power']=train_df['Power'].astype(float)
```

```python
train_df.dtypes
```

```python
#integer replacement(powwer,mileage,engine of a vehicle must be vale greater than 0, so we treat 0 as missing value)
#consider 0 as missing value
train_df.loc[train_df.Engine==0,'Engine']=np.NaN
train_df.loc[train_df.Mileage==0,'Mileage']=np.NaN
train_df.loc[train_df.Power==0,'Power']=np.NaN
```

```python
train_df.isna().sum()
```

```python
train_df['Mileage'].unique()
```

```python
a=train_df['Mileage'].mean()
train_df['Mileage']=train_df['Mileage'].fillna(a)
train_df.isna().sum()
```
