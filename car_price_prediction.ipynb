```python
df['Brand']=df['Name'].apply(lambda x:x.split()[0])
```

```python
df['Brand']
```

```python
df['Brand'].value_counts()
```

```python
df.drop('Name',inplace=True,axis=1)
```

```python
df['Mileage']=df['Mileage'].str.replace('km/kg','')
df['Mileage']=df['Mileage'].str.replace('kmpl','')
```

```python
df['Engine']=df['Engine'].str.replace('CC','')
df['Power']=df['Power'].str.replace('bhp','')
```

```python
df['Engine']=df['Engine'].str.replace('null','0')
df['Mileage']=df['Mileage'].str.replace('null','0')
df['Power']=df['Power'].str.replace('null','0')
df
```

```python
df.loc[df.Engine==0,'Engine']=np.NaN
df.loc[df.Mileage==0,'Mileage']=np.NaN
df.loc[df.Power==0,'Power']=np.NaN
```

```python
df['Mileage']=df['Mileage'].astype(float)
df['Engine']=df['Engine'].astype(float)
df['Power']=df['Power'].astype(float)
```

```python
df.isna().sum()
```

```python
df.drop(['Unnamed: 0','New_Price'],axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
x=df.drop(['Price'],axis=1)
x
```

```python
y=df['Price']
y
```

```python
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
num_pipe=Pipeline(steps=
                  [('fillmissing',SimpleImputer(strategy='mean')),
                   ('scale',StandardScaler())])
```

```python
cat_pipe=Pipeline(steps=
                  [('fillmissing',SimpleImputer(strategy='most_frequent')),
                   ('onehotencoder',OneHotEncoder(handle_unknown='ignore',sparse_output=False))])
```

```python
x.dtypes
```

```python
x.columns
```

```python
numeric_columns=['Year', 'Kilometers_Driven','Mileage', 'Engine', 'Power']
cat_column=['Location','Fuel_Type', 'Transmission','Owner_Type','Seats','Brand']
```

```python
col_trans=ColumnTransformer(transformers=[
    ('numeric',num_pipe,numeric_columns),
    ('category',cat_pipe,cat_column)
],remainder="passthrough",n_jobs=-1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=make_pipeline(col_trans,LinearRegression())
```

```python
model
```

```python
model.fit(x,y)
```
